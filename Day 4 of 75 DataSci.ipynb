```python
import pandas as pd
import matplotlib.pyplot as plt
#import seaborn as sns
```

```python
df = pd.read_csv('sales_data.csv')

# Display the first few rows of the dataset to inspect it
print(df.head())
```

```python
# Check for missing values
print(df.isnull().sum())

# Replace missing values in 'Sales' column with the mean of the column
df['Sales'].fillna(df['Sales'].mean(), inplace=True)
```

```python
total_sales_per_region = df.groupby('Date')['Sales'].sum().reset_index()

# Plot using Matplotlib
plt.figure(figsize=(10, 6))

# Create a bar plot
plt.bar(total_sales_per_region['Date'], total_sales_per_region['Sales'], color='skyblue')

# Add labels and title
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.title('Total Sales per Region')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability

# Show the plot
plt.show()
```
