```python
import matplotlib.pyplot as plt
import numpy as np

# Define the linear function parameters
m = 2  # slope
b = 3  # y-intercept

# Generate x values (e.g., from -10 to 10)
x = np.linspace(-10, 10, 100)
# Calculate y values based on the linear function
y = m * x + b

# Plot the line
plt.plot(x, y, label=f"y = {m}x + {b}", color="blue")

# Customize the plot
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Graph of the Linear Function y = mx + b")
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
