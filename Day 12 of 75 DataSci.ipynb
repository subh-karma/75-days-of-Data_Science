# Problem:
We'll use a synthetic dataset where X represents the marketing spend, and Y represents profit. We'll create a scatter plot of the data points and plot the regression line to see the fit.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Step 1: Generate Sample Data
# Marketing spend (in thousands)
X = np.array([10, 15, 20, 25, 30, 35, 40]).reshape(-1, 1)
# Profit (in thousands)
Y = np.array([12, 18, 25, 30, 35, 45, 50])

# Step 2: Fit a Linear Regression Model
model = LinearRegression()
model.fit(X, Y)
Y_pred = model.predict(X)

# Step 3: Calculate the Mean Squared Error (Optional)
mse = mean_squared_error(Y, Y_pred)
print(f"Mean Squared Error: {mse}")

# Step 4: Plot the Data Points and Regression Line
plt.figure(figsize=(8, 6))
plt.scatter(X, Y, color='blue', label='Actual Data')
plt.plot(X, Y_pred, color='red', label='Regression Line')
plt.xlabel('Marketing Spend (in thousands)')
plt.ylabel('Profit (in thousands)')
plt.title('Linear Regression: Marketing Spend vs Profit')
plt.legend()
plt.show()
```
